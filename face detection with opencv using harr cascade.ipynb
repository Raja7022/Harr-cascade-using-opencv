# Face Detection with opencv using harr cascade

```python
#specify the  xml path for cascade classifer
face_cascade=cv2.CascadeClassifier('G:\\haar cascade classifier\\data\\haarcascades xml features\\haarcascade_frontalface_default.xml')

#specify video source
cap=cv2.VideoCapture(0)

while True:
    
    #capture FRAME by FRAME
    ret,img=cap.read()
    
    # convert into gray
    gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
    faces=face_cascade.detectMultiScale(gray,scaleFactor=1.5,minNeighbors=5)
    
    #loop through the faces and print the values
    for (x,y,w,h) in faces:
        print(x,y,w,h)
        
        # region of interest(ROI) 
        roi_gray=gray[y:y+h,x:x+w]
        roi_color=img[y:y+h,x:x+w]
        
        # save gray region of interest 
        img_item="myyy hamsa.png"
        cv2.imwrite(img_item,roi_gray)
        
        #draw rectangle around face
        color=(255,0,0)
        stroke=2
        end_cord_x=x+w
        end_cord_y=y+h
        cv2.rectangle(img,(x,y),(end_cord_x,end_cord_y),color,stroke)
    
    # Displayy resulting frame
    cv2.imshow("my video source",img)
    if cv2.waitKey(30) & 0xff==ord('q'):
        break
        
#when everything is done release video capture        
cap.release()
cv2.destroyAllWindows()
```
